```python
engagement_dt = pd.to_datetime(engagement['time_stamp'])
engagement['date'] = engagement_dt.dt.date
engagement.info()
```

```python
user_count = engagement.groupby('user_id').count()
user_3x = user_count[user_count['date'] >=3]
user_3x
engagement['user_3x'] = engagement['user_id'].apply(lambda x: 1 if x in user_3x.index else 0)

engagement
```

```python
engagement_3x = engagement[engagement['user_3x'] == 1]
engagement_3x.head(20)
```

```python
engagement_3x.groupby('user_id').count()
```

There are 2,248 users who have visited on 3 separate occasions.  Next, to find out how many of those users made 3 visits within a 7 day stretch.

```python
engagement_3x.info()
```

```python
engagement_3x['date_num'] = pd.to_datetime(engagement_3x['date']).dt.strftime("%Y%m%d")
engagement_3x['date_num'] = pd.to_datetime(engagement_3x['date_num'], format='%Y%m%d')

```

```python
# engagement_3x.groupby('date')['user_id'].rolling(7).count().head(30)

engagement_3x = engagement_3x.set_index('date_num')
engagement_3x.groupby('date_num')['user_id'].resample('W').sum()
```
